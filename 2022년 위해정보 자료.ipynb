```python
import pandas as pd
```

```python
df = pd.read_csv("한국소비자원 소비자 위해 정보_20220630.csv", encoding = 'CP949')
df
```

```python
# 총 66개
df.query("품목소분류 == '유모차'")
```

```python
# 위해자 연령 평균연령 8.94세
df1 = df.query('품목소분류 == "유모차"')
df1.groupby('품목소분류').agg(mean_agg = ('위해자연령', 'mean'))
```

```python
# 위험종류별 횟수
df2 = df.query('품목소분류 == "유모차"').groupby('위험 및 위해원인').agg(count_agg = ('위해자연령', 'count'))
df2.sort_values('count_agg', ascending=False)
```

```python
# 유모차 추락 52개 증명
df3 = df2.query('위험 및 위해원인 == "추락"')
df3.shape
```

```python
import matplotlib.pyplot as plt

ratio = [52, 5, 3, 1, 1, 1, 1, 1]
labels = ['Fall', 'Other Defects and Failures', 'Other Risks and Hazards Contents',
          'Other physical shocks', 'Other Chemicals', 'Pressing·getting stuck', 
         'Slips·falls', 'Rupture·breakage·bend']

plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.legend(loc = (1, 0.5))
plt.show()
```

```python
# 부채꼴 스타일 지정하기
import matplotlib.pyplot as plt

ratio = [52, 5, 3, 1, 1, 1, 1, 1]
labels = ['Fall', 'Other Defects and Failures', 'Other Risks and Hazards Contents',
          'Other physical shocks', 'Other Chemicals', 'Pressing·getting stuck', 
         'Slips·falls', 'Rupture·breakage·bend']
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}


plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, wedgeprops=wedgeprops)
plt.legend(loc = (1, 0.5))
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정
plt.rcParams['font.family'] = 'NanumGothic'

plt.figure(figsize=(12, 30))
sns.countplot(y=df2['count_agg'], order=df2['count_agg'].value_counts().index)
plt.yticks(fontsize=12)
plt.title('Frequnecy')
plt.show()
```
