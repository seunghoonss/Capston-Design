```python
import pandas as pd
```

```python
#데이터 166746개 (2020~2022.3)
df = pd.read_csv("1한국소비자원 소비자 위해 정보_20220630.csv", encoding = 'CP949')
df
```

```python
# 총 476개
df.query("품목소분류 == '유모차'")
```

```python
# 위해자 연령 평균연령 13.66세
df1 = df.query('품목소분류 == "유모차"')
df1.groupby('품목소분류').agg(mean_agg = ('위해자연령', 'mean'))
```

```python
# 위험종류별 횟수 #12개
df2 = df.query('품목소분류 == "유모차"').groupby('위험 및 위해원인').agg(count_agg = ('위해자연령', 'count'))
df2.sort_values('count_agg', ascending=False)
```

```python
df2['count_agg'].plot.hist()
```
