```python
import pandas as pd
```

```python
#데이터 73011개 (2019)
df = pd.read_csv("부산 경사도 구역통계.csv", encoding = 'CP949')
df
```

```python
df.groupby('ADM_DR_NM').agg( mean_agg = ('_mean', 'mean'),min_agg = ('_min', 'mean'),
                            max_agg = ('_max', 'mean'), range_agg = ('_range', 'mean'))
```

```python
df.mean().to_frame()
```
